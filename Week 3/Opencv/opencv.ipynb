```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Bài 1

```python
abs_path = []
rel_path = []
for each in os.listdir('image'):
    abs_p = os.path.join(os.path.abspath('image'), each)
    abs_path.append(abs_p)
    rel_path.append(os.path.join('image', each))
print(abs_path)
print(rel_path)
```

# Bài 2.1:

```python
img = cv.imread(rel_path[1])
plt.imshow(img)
plt.show()
```

```python
b_channel, g_channel, r_channel = img.copy(), img.copy(), img.copy()
```

```python
r_channel[:, :, 0:2] = 0
```

```python
g_channel[:, :, 0] = 0
g_channel[:, :, 2] = 0
```

```python
b_channel[:, :, 1:3] = 0
```

```python
plt.subplot(1, 3, 1)
plt.imshow(cv.cvtColor(b_channel, cv.COLOR_BGR2RGB))
plt.subplot(1, 3, 2)
plt.imshow(cv.cvtColor(g_channel, cv.COLOR_BGR2RGB))
plt.subplot(1, 3, 3)
plt.imshow(cv.cvtColor(r_channel, cv.COLOR_BGR2RGB))
plt.show()
```

# Bài 2.2

```python
img = cv.imread(rel_path[2])
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
histogram = np.zeros((256, ))
```

```python
for i in gray:
    for j in i:
        histogram[j] += 1
```

```python
plt.plot(histogram)
plt.show()
```
