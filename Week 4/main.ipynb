```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
def acc(result, ans):
    rs = ans - result
    print(result)
    print(ans)
    print('Wrong labels: {}/{}'.format(sum(np.absolute(rs)), len(result)))
```

```python
data = pd.read_csv('data.csv', header=None)
X = np.array(data.iloc[:, [0, 1]])
y = np.array(data.iloc[:, 2])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.show()
```

# Kernel SVM

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', gamma='scale', C=0.5)
model.fit(X_train, y_train)
```

```python
result1 = model.predict(X_test)
```

# PLA

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(power_t=0.5)
model.fit(X_train, y_train)
```

```python
result2 = model.predict(X_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
result3 = model.predict(X_test)
```

# Result

## Kernel SVM

```python
acc(result1, y_test)
```

## PLA

```python
acc(result2, y_test)
```

## LogisticRegression

```python
acc(result3, y_test)
```
